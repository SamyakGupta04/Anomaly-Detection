```python
import random
import os
import pandas as pd
import time

# Set base path inside your mounted Drive
base_path = '/walmart/drive/MyDrive/attacks'

# Column headers
main_labels = ["Flow ID", "Source IP", "Source Port", "Destination IP", "Destination Port", "Protocol", "Timestamp",
               "Flow Duration", "Total Fwd Packets", "Total Backward Packets", "Total Length of Fwd Packets",
               "Total Length of Bwd Packets", "Fwd Packet Length Max", "Fwd Packet Length Min", "Fwd Packet Length Mean",
               "Fwd Packet Length Std", "Bwd Packet Length Max", "Bwd Packet Length Min", "Bwd Packet Length Mean",
               "Bwd Packet Length Std", "Flow Bytes/s", "Flow Packets/s", "Flow IAT Mean", "Flow IAT Std", "Flow IAT Max",
               "Flow IAT Min", "Fwd IAT Total", "Fwd IAT Mean", "Fwd IAT Std", "Fwd IAT Max", "Fwd IAT Min",
               "Bwd IAT Total", "Bwd IAT Mean", "Bwd IAT Std", "Bwd IAT Max", "Bwd IAT Min", "Fwd PSH Flags",
               "Bwd PSH Flags", "Fwd URG Flags", "Bwd URG Flags", "Fwd Header Length", "Bwd Header Length",
               "Fwd Packets/s", "Bwd Packets/s", "Min Packet Length", "Max Packet Length", "Packet Length Mean",
               "Packet Length Std", "Packet Length Variance", "FIN Flag Count", "SYN Flag Count", "RST Flag Count",
               "PSH Flag Count", "ACK Flag Count", "URG Flag Count", "CWE Flag Count", "ECE Flag Count", "Down/Up Ratio",
               "Average Packet Size", "Avg Fwd Segment Size", "Avg Bwd Segment Size", "Fwd Avg Bytes/Bulk",
               "Fwd Avg Packets/Bulk", "Fwd Avg Bulk Rate", "Bwd Avg Bytes/Bulk", "Bwd Avg Packets/Bulk",
               "Bwd Avg Bulk Rate", "Subflow Fwd Packets", "Subflow Fwd Bytes", "Subflow Bwd Packets", "Subflow Bwd Bytes",
               "Init_Win_bytes_forward", "Init_Win_bytes_backward", "act_data_pkt_fwd", "min_seg_size_forward",
               "Active Mean", "Active Std", "Active Max", "Active Min", "Idle Mean", "Idle Std", "Idle Max", "Idle Min",
               "Label", "External IP"]
main_labels = ",".join(main_labels)

attacks = ["BENIGN", "Bot", "DDoS", "DoS GoldenEye", "DoS Hulk", "DoS Slowhttptest", "DoS slowloris",
           "FTP-Patator", "Heartbleed", "Infiltration", "PortScan", "SSH-Patator", "Web Attack - Brute Force",
           "Web Attack - Sql Injection", "Web Attack - XSS"]

benign = 2359289

dict_attack = {
    "Bot": 1966,
    "DDoS": 41835,
    "DoS GoldenEye": 10293,
    "DoS Hulk": 231073,
    "DoS Slowhttptest": 5499,
    "DoS slowloris": 5796,
    "FTP-Patator": 7938,
    "Heartbleed": 11,
    "Infiltration": 36,
    "PortScan": 158930,
    "SSH-Patator": 5897,
    "Web Attack - Brute Force": 1507,
    "Web Attack - XSS": 652,
    "Web Attack - Sql Injection": 21
}

all_data_path = '/walmart/drive/My Drive/walmart/all_data.csv'

seconds = time.time()

for attack_type in dict_attack:
    benign_count, attack_count = 0, 0
    output_file = os.path.join(base_path, f"{attack_type}.csv")
    benign_num = int(benign / (dict_attack[attack_type] * (7 / 3)))

    with open(output_file, "w") as ths:
        ths.write(main_labels + "\n")

        with open(all_data_path, "r") as file:
            while True:
                try:
                    line = file.readline()
                    if not line:
                        break
                    line = line.rstrip('\n')
                    k = line.split(",")
                    if len(k) < 84:
                        continue  # skip incomplete lines
                    if k[83] == "BENIGN":
                        if random.randint(1, benign_num) == 1:
                            ths.write(line + "\n")
                            benign_count += 1
                    elif k[83] == attack_type:
                        ths.write(line + "\n")
                        attack_count += 1
                except:
                    break

    print(f"{attack_type} file completed\n attack: {attack_count}\n benign: {benign_count}\n\n")

# Merge all Web Attack files into one
web_attacks = ["Web Attack - Brute Force", "Web Attack - XSS", "Web Attack - Sql Injection"]
merged_web_file = os.path.join(base_path, "Web Attack.csv")

flag = True
for web_attack in web_attacks:
    file_path = os.path.join(base_path, f"{web_attack}.csv")
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        if flag:
            df.to_csv(merged_web_file, index=False)
            flag = False
        else:
            df.to_csv(merged_web_file, index=False, header=False, mode="a")
        os.remove(file_path)

print("Mission accomplished!")
print("Operation time:", time.time() - seconds, "seconds")

```
